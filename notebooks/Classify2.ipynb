```python
def periodize(year,ybin=20):
    y = year // ybin * ybin
    return f'{y}-{y+ybin}'

df_meta['decade'] = df_meta.year.apply(lambda x: x // 10 * 10)
df_meta['period'] = df_meta.year.map(periodize)
```
