```python
def summarize_sentence_stats(doc):
    df = pd.DataFrame(get_tree_stats(sent) for sent in doc.sentences)
    s = df.mean(numeric_only=True)
    s.index = ['sent_'+str(i) for i in s.index]
    return dict(s)


summarize_sentence_stats(doc)
```

```python


def get_sent_feat_counts(doc, feat2word2count=None, feat2word2eg=None, context_len=CONTEXT_LEN):
    return [get_tree_stats(sent) for sent in doc.sentences]

def gen_sent_feat_counts(id, force=False):
    if not force and id in STASH_SENT_FEAT_COUNTS:
        return STASH_SENT_FEAT_COUNTS[id]
    
    docstr = input_stash[id]
    doc = stanza.Document.from_serialized(docstr)
    counts = get_sent_feat_counts(doc)
    STASH_SENT_FEAT_COUNTS[id] = counts
    return counts
```
