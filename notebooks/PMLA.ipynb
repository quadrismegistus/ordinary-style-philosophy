```python
fn = '../data/raw/LitStudiesJSTOR.jsonl'
total = 71902

from tqdm import tqdm
import orjsonl
import pandas as pd
```

```python
def iter_jsonl(fn, total=total):
    yield from tqdm(orjsonl.stream(fn), total=total)
```

```python
df=pd.DataFrame(iter_jsonl(fn))
```

```python
len(df)
```

```python
df['year'] = df['datePublished'].apply(lambda x: x.split('-')[0]).apply(int)
df['decade'] = df['year'] // 10 * 10
```
