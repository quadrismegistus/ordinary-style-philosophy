```python
# from bounter import bounter
import json
PATH_TOTAL_TEXT_COUNTS = os.path.join(PATH_DATA, 'total_text_counts.json')

def get_total_text_counts(path=PATH_TOTAL_TEXT_COUNTS, force=False):
    if not force and os.path.exists(path):
        try:
            return json.load(open(path))
        except:
            pass
    text_counts = Counter()
    for id,counts in tqdm(STASH_COUNTS.items(),total=len(STASH_COUNTS)):
        if counts:
            for word, count in counts.items():
                text_counts[word] += 1
    out = dict(text_counts.most_common())
    with open(path, 'w') as f:
        json.dump(out, f)
    return out

# get_total_text_counts(force=False)
```

```python
def is_stopword(word):
    # return word in STOP_WORDS
    return len(word)<4

def get_top_words(n=10, remove_stopwords=True,except_first_n=100):
    d = get_total_text_counts()
    d = Counter({k:v for k,v in d.items() if not is_stopword(k)})
    out = [w for w,c in d.most_common(n+except_first_n)]
    if except_first_n and len(out)>except_first_n:
        out = out[except_first_n:]
    return out

pprint(' '.join(get_top_words()))
```

```python
def get_word_group_counts(words_list):
    d = get_total_text_counts()
    out = []
    for i,(id,counts) in enumerate(tqdm(STASH_COUNTS.items(),total=len(STASH_COUNTS))):
        if not counts:
            continue
        total = sum(counts.values())
        group_counts = {c:v for c,v in counts.items() if c in words_list}
        total_group = sum(group_counts.values())
        outd = {'id':id}#, 'num_words_group':total_group, 'num_words_text':total}
        for c,v in group_counts.items():
            outd[c] = v/total_group
        out.append(outd)
        # if i>10:
            # break
    return pd.DataFrame(out).set_index('id')

```

```python
def classify_data(data):

    from sklearn.linear_model import LogisticRegression
    from sklearn.metrics import classification_report, accuracy_score

    # Ensure data is clean (fill NaNs)
    _data = data.copy()
    _data['_target'] = target = data.index.str.split('/').str[0]

    min_target_size = min(target.value_counts())
    _data = _data.groupby('_target').sample(n=min_target_size)
    X_data = _data.drop(columns=['_target']).fillna(0).values
    y_data = _data['_target'].values

    # Use the train/test split (80/20 is appropriate for 57k samples)
    from sklearn.model_selection import train_test_split
    X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size=0.2, random_state=42)

    # Initialize Logistic Regression
    model = LogisticRegression(class_weight='balanced', max_iter=1000)

    print(f"Training on {len(X_train)} samples, testing on {len(X_test)} samples...")
    model.fit(X_train, y_train)

    # Predictions
    y_pred = model.predict(X_test)

    print("\nClassifier Results (on Test Set):")
    print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}")
    print(classification_report(y_test, y_pred))
    
```

```python
classify_data(data)
```
