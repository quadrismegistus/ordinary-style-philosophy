```python
def parse_for_keyword(ids,keyword):
    for id in tqdm(ids):
        txt = get_corpus_txt(id)
        sents = txt.split('\n')
        for sent in sents:
            doc = get_nlp_doc(sent, id=id)
            if keyword in doc.text:
                print(id,sent)
                break
```
